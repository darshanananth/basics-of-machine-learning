```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from scipy.stats import bernoulli, binom, poisson, uniform, norm
import pandas as pd

# Probability Distribution Experiment

# Throwing a fair die multiple times
dice = [x for x in range(1, 7)]

# Default number of trials = 10
def throw_dice(n=10):
    r = []  # list of random results
    for _ in range(n):
        value = random.choice(dice)
        r.append(value)
    return np.array(r)  # function returns array results

# Example calls to the function
print("Throwing the fair die 10 times:", throw_dice())
print("Throwing the fair die 20 times:", throw_dice(n=20))

# Displaying averages for different numbers of trials
for i in [1, 5, 10, 50, 100, 500, 1000, 5000, 10000]:
    print("Average of {} dice throws: {}".format(i, round(throw_dice(i).mean(), 2)))

# Discrete Probability Distribution

# Bernoulli Distribution
print("\nBernoulli Distribution - Tossing a loaded coin")
print("Loaded towards tails (p=0.2):", bernoulli.rvs(p=0.2, size=10))
print("Loaded towards heads (p=0.7):", bernoulli.rvs(p=0.7, size=10))

# Bernoulli with Multiple trails
N_trials = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000]
prob = 0.5
average = []

for i in N_trials:
    for n in range(1, i + 1):
        average.append(np.mean(bernoulli.rvs(p=prob, size=10)))

    plt.title("Distribution with {} trials of 10 coin tosses".format(i))
    plt.hist(average, bins=10 if i == 10 else 25, edgecolor='deeppink', color='pink')
    plt.xlim(0.0, 1.0)
    plt.xticks([0.1 * i for i in range(11)])
    plt.grid(True)
    plt.show()

# Binomial Distribution
print("\nBinomial Distribution - 8 coins flipped 10 times with p=0.25")
k = binom.rvs(8, 0.25, size=10)
print("Number of successes for each trial:", k)
print("Average of the successes:", np.mean(k))

# Poisson Distribution
print("\nPoisson Distribution")
la_values = [0.5, 1, 5]

plt.figure(figsize=(8, 6))
for la in la_values:
    rv = poisson(la)
    x = np.arange(0, 20, 1)
    pmf = rv.pmf(x)
    plt.plot(x, pmf, label=f'Î»={la}')

plt.legend()
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()

# Uniform (Continuous) Distribution
print("\nUniform (Continuous) Distribution")
print("Random continuous variable between 0 and 1:", uniform.rvs(size=6))
print("Random floats between -10 and 0:", uniform.rvs(loc=-10, scale=10, size=6))
print("Random floats between -30 and -30+scale (20):", uniform.rvs(loc=-30, scale=20, size=6))

# Visualizing Uniform Distribution
plt.figure(figsize=(8, 6))
samples_uniform = uniform.rvs(size=1000)
plt.hist(samples_uniform, bins=30, density=True, edgecolor='black', color='skyblue', alpha=0.7)
plt.title('Uniform Distribution')
plt.xlabel('Random Variable')
plt.ylabel('Probability Density')
plt.show()
```
